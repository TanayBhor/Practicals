```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the original image
original_image = cv2.imread('ip.jpg', cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

# Subsample the image to different sizes
subsampled_images = []
sizes = [(512, 512), (256, 256), (128, 128), (64, 64), (32, 32)]

for i, size in enumerate(sizes):
    subsampled_image = cv2.resize(original_image, size, interpolation=cv2.INTER_NEAREST)
    subsampled_images.append(subsampled_image)

    # Display subsampled images
    plt.subplot(2, 3, i + 2)
    plt.imshow(subsampled_image, cmap='gray')
    plt.title(f'Subsampled {size[0]}x{size[1]}')

plt.tight_layout()
plt.show()
```

```python
# Resample the images back to the original size
resampled_images = []

for i, size in enumerate(sizes):
    resampled_image = cv2.resize(subsampled_images[i], (1024, 1024), interpolation=cv2.INTER_LINEAR)
    resampled_images.append(resampled_image)

# Display resampled images
plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

for i, size in enumerate(sizes):
    plt.subplot(2, 3, i + 2)
    plt.imshow(resampled_images[i], cmap='gray')
    plt.title(f'Resampled {size[0]}x{size[1]}')

plt.tight_layout()
plt.show()
```

```python
# Display histogram, equalized histogram, and image with equalized histogram
plt.figure(figsize=(15, 5))

# Original image histogram
plt.subplot(1, 3, 1)
plt.hist(original_image.flatten(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Original Image Histogram')

# Equalize histogram
equalized_image = cv2.equalizeHist(original_image)

# Equalized histogram
plt.subplot(1, 3, 2)
plt.hist(equalized_image.flatten(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Equalized Image Histogram')

# Image with equalized histogram
plt.subplot(1, 3, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Image with Equalized Histogram')

plt.tight_layout()
plt.show()
```
