```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image_path = 'flower.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

# Calculate and display the original image histogram
plt.subplot(1, 3, 2)
plt.hist(original_image.flatten(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Original Image Histogram')

# Equalize the image
equalized_image = cv2.equalizeHist(original_image)

# Calculate and display the equalized image histogram
plt.subplot(1, 3, 3)
plt.hist(equalized_image.flatten(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Equalized Image Histogram')

# Display the equalized image
plt.figure(figsize=(4, 4))
plt.imshow(equalized_image, cmap='gray')
plt.title('Image with Equalized Histogram')

plt.tight_layout()
plt.show()
```
