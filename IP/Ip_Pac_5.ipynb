```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image_path = 'horse.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Sobel edge detection
sobel_x = cv2.Sobel(original_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(original_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = np.sqrt(sobel_x**2 + sobel_y**2)

# Apply Canny edge detection
canny_edges = cv2.Canny(original_image, 50, 150)

# Apply Laplacian edge detection
laplacian_edges = cv2.Laplacian(original_image, cv2.CV_64F)

# Display the results
plt.figure(figsize=(12, 8))

# Original image
plt.subplot(2, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

# Sobel edges
plt.subplot(2, 3, 2)
plt.imshow(sobel_edges, cmap='gray')
plt.title('Sobel Edge Detection')

# Canny edges
plt.subplot(2, 3, 3)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')

# Laplacian edges
plt.subplot(2, 3, 4)
plt.imshow(laplacian_edges, cmap='gray')
plt.title('Laplacian Edge Detection')

plt.tight_layout()
plt.show()
```
